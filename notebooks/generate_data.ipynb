##### Import Data

```python
from datasets import load_dataset

dataset = load_dataset("wikitext", 'wikitext-103-raw-v1')
dataset
```

#### Process Inference

```python
import json
import pandas as pd
from sgarbi import QABuilderPipeline

producer = QABuilderPipeline()

df = pd.DataFrame(dataset['train'][:1000])
# shuffle
df = df.sample(frac=1).reset_index(drop=True)
df = df[:40]
for i, row in df.iterrows():
    input_text = row['text']
    output = producer(input_text, similarity_threshold=0.8, json_output=True, silent_mode=True)
    
    print(output)
    print()
```
